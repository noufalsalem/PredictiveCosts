```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('data.xlsx')
```

```python
df.head()
```

```python
df.describe().T
```

```python
df.info()
```

```python
new_df = df.copy()
```

```python
new_df.index = df.Month
```

```python
new_df.drop('Month',axis = 1 , inplace = True)
```

```python
sns.histplot(new_df['National Cement Black (50 Kg)'])
```

```python
new_df['National Cement Black (50 Kg)'].plot()
```

```python
for col in new_df.columns:
    plt.figure(figsize = (10,4))
    
    plt.subplot(121)
    plt.xlabel('Month')
    plt.ylabel(col)
    plt.plot(new_df[col])
    
    plt.subplot(122)
    sns.histplot(new_df[col])
    plt.show()
```

```python
#### Box-plot and Violin plot

for col in new_df.columns:
    plt.figure(figsize = (10,4))
    
    plt.subplot(121)
    sns.boxplot(x = new_df[col])
    
    plt.subplot(122)
    sns.violinplot(x = new_df[col])
    print('\n')
    plt.show()
```

```python
### create dataset for modeling


new_df['National Cement Black (50 Kg)'].reset_index().to_csv('national_cement_black_50kg.csv',index = False)
```
